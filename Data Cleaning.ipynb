### Unique value counts of training data and test data

```python
df_train_unique = df_train.agg(['nunique']).transpose().reset_index()
df_train_unique.columns = ['column', 'unique_value_count']
```

```python
display(df_train_unique)
```

```python
drop_cols_df = df_train_unique[df_train_unique.unique_value_count.isin([0, 1])]

display(drop_cols_df)
```

```python
to_drop_cols = drop_cols_df.column.values.tolist()

df_train = df_train.drop(columns=to_drop_cols)
```

```python
df_train = df_train.drop(columns=['geodimid_cr'])
```

```python
def missing_value_stats(df):
    num_rows = df.shape[0]
    # Missing value count by columns
    df_missing = df.isna().sum().reset_index()
    # Renaming columns
    df_missing.columns = ['column', 'missing_values']
    # Getting only those columns having missing values > 0
    df_missing = df_missing[df_missing['missing_values'] > 0]
    df_missing.reset_index(drop=True, inplace=True)
    # Calculating percentage of missing
    df_missing['missing_percentage'] = df_missing['missing_values'].apply(
        lambda missing_value: round((missing_value / num_rows) * 100, 2)
    )
    return df_missing
```

```python
df_train_missing = missing_value_stats(df_train)

display(df_train_missing[df_train_missing.missing_percentage > 80].sort_values(by=['missing_percentage']))
```

```python
missing_cols = df_train_missing.loc[df_train_missing.missing_percentage > 90, 'column'].values.tolist()

df_train = df_train.drop(columns=missing_cols)
```
